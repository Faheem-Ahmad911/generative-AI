```python
"""
Generate PDF report for Prompt Engineering Assignment
Student: BSDSF23M025 | Course: Generative AI
"""

from fpdf import FPDF
from fpdf.enums import XPos, YPos

OUTPUT_PATH = "d:/generative-AI/PromptEngineering/BSDSF23M025_Report.pdf"

# ─── PDF Class ────────────────────────────────────────────────────────────────
class Report(FPDF):
    def __init__(self):
        super().__init__(orientation="P", unit="mm", format="A4")
        self.set_auto_page_break(auto=True, margin=20)
        self.set_margins(20, 20, 20)

    def header(self):
        if self.page_no() == 1:
            return
        self.set_font("Helvetica", "I", 8)
        self.set_text_color(120, 120, 120)
        self.cell(0, 6, "Prompt Engineering & LLM Behavior Analysis  |  BSDSF23M025",
                  align="L", new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_draw_color(200, 200, 200)
        self.line(20, self.get_y(), 190, self.get_y())
        self.ln(2)
        self.set_text_color(0, 0, 0)

    def footer(self):
        self.set_y(-15)
        self.set_font("Helvetica", "I", 8)
        self.set_text_color(120, 120, 120)
        self.cell(0, 10, f"Page {self.page_no()}", align="C")
        self.set_text_color(0, 0, 0)

    def title_line(self, text, size=16, style="B", color=(0, 51, 102), spacing=6):
        self.set_font("Helvetica", style, size)
        self.set_text_color(*color)
        self.multi_cell(0, spacing, text, new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_text_color(0, 0, 0)

    def section(self, number, title):
        self.ln(4)
        self.set_fill_color(0, 51, 102)
        self.set_text_color(255, 255, 255)
        self.set_font("Helvetica", "B", 11)
        self.cell(0, 8, f"  {number}  {title}", fill=True,
                  new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_text_color(0, 0, 0)
        self.ln(2)

    def subsection(self, title):
        self.ln(2)
        self.set_font("Helvetica", "B", 10)
        self.set_text_color(0, 51, 102)
        self.multi_cell(0, 6, title, new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_text_color(0, 0, 0)
        self.set_draw_color(0, 51, 102)
        self.line(20, self.get_y(), 100, self.get_y())
        self.ln(2)

    def body(self, text, size=9.5):
        self.set_font("Helvetica", "", size)
        self.set_text_color(30, 30, 30)
        self.multi_cell(0, 5.5, text, new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_text_color(0, 0, 0)
        self.ln(1)

    def bullet(self, text, indent=5):
        self.set_font("Helvetica", "", 9.5)
        self.set_x(20 + indent)
        self.set_text_color(30, 30, 30)
        self.multi_cell(0, 5.5, f"\u2022  {text}", new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_text_color(0, 0, 0)

    def code_block(self, text):
        self.ln(1)
        self.set_fill_color(245, 245, 245)
        self.set_draw_color(200, 200, 200)
        self.set_font("Courier", "", 8)
        self.set_text_color(30, 30, 30)
        self.multi_cell(0, 4.8, text, border=1, fill=True,
                        new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_font("Helvetica", "", 9.5)
        self.set_text_color(0, 0, 0)
        self.ln(1)

    def response_block(self, label, text):
        self.ln(1)
        self.set_font("Helvetica", "B", 8.5)
        self.set_text_color(80, 80, 80)
        self.cell(0, 5, label, new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_fill_color(250, 248, 240)
        self.set_draw_color(200, 180, 140)
        self.set_font("Helvetica", "I", 8.5)
        self.set_text_color(50, 50, 50)
        self.multi_cell(0, 5, text, border=1, fill=True,
                        new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_text_color(0, 0, 0)
        self.ln(1)

    def table(self, headers, rows, col_widths=None):
        self.ln(1)
        total_w = 170
        if not col_widths:
            col_widths = [total_w // len(headers)] * len(headers)
        # Header row
        self.set_fill_color(0, 51, 102)
        self.set_text_color(255, 255, 255)
        self.set_font("Helvetica", "B", 8)
        for h, w in zip(headers, col_widths):
            self.cell(w, 6, h, border=1, fill=True, align="C")
        self.ln()
        # Data rows
        self.set_text_color(30, 30, 30)
        for i, row in enumerate(rows):
            fill = i % 2 == 0
            self.set_fill_color(235, 242, 250) if fill else self.set_fill_color(255, 255, 255)
            self.set_font("Helvetica", "", 7.5)
            for cell_val, w in zip(row, col_widths):
                self.cell(w, 5.5, str(cell_val), border=1, fill=True, align="C")
            self.ln()
        self.set_text_color(0, 0, 0)
        self.ln(2)

    def callout(self, label, text, bg=(240, 248, 255), border_color=(0, 100, 180)):
        self.ln(1)
        self.set_fill_color(*bg)
        self.set_draw_color(*border_color)
        self.set_font("Helvetica", "B", 8.5)
        self.set_text_color(*border_color)
        self.cell(0, 5.5, f"  {label}", border="LTR", fill=True,
                  new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_font("Helvetica", "", 8.5)
        self.set_text_color(30, 30, 30)
        self.multi_cell(0, 5, f"  {text}", border="LBR", fill=True,
                        new_x=XPos.LMARGIN, new_y=YPos.NEXT)
        self.set_text_color(0, 0, 0)
        self.ln(1)

# ─── Build PDF ─────────────────────────────────────────────────────────────────
pdf = Report()

# ═══════════════════════════════════════════════════════════════════════════════
# COVER PAGE
# ═══════════════════════════════════════════════════════════════════════════════
pdf.add_page()
pdf.ln(25)
pdf.set_fill_color(0, 51, 102)
pdf.rect(0, 60, 210, 70, "F")
pdf.set_y(70)
pdf.set_font("Helvetica", "B", 22)
pdf.set_text_color(255, 255, 255)
pdf.multi_cell(0, 10, "Prompt Engineering &\nLLM Behavior Analysis", align="C",
               new_x=XPos.LMARGIN, new_y=YPos.NEXT)
pdf.ln(3)
pdf.set_font("Helvetica", "", 12)
pdf.cell(0, 8, "Generative AI — Assignment Report", align="C",
         new_x=XPos.LMARGIN, new_y=YPos.NEXT)
pdf.set_text_color(0, 0, 0)
pdf.set_y(155)

meta = [
    ("Student ID",    "BSDSF23M025"),
    ("Course",        "Generative AI"),
    ("Model Used",    "LLaMA 3.1 8B Instant"),
    ("API Provider",  "Groq (OpenAI-compatible endpoint)"),
    ("Submission",    "March 2025"),
]
pdf.set_font("Helvetica", "", 10)
for label, value in meta:
    pdf.set_font("Helvetica", "B", 10)
    pdf.set_text_color(0, 51, 102)
    pdf.cell(50, 7, f"{label}:", new_x=XPos.RIGHT, new_y=YPos.LAST)
    pdf.set_font("Helvetica", "", 10)
    pdf.set_text_color(30, 30, 30)
    pdf.cell(0, 7, value, new_x=XPos.LMARGIN, new_y=YPos.NEXT)
pdf.set_text_color(0, 0, 0)

pdf.ln(10)
pdf.set_font("Helvetica", "B", 9)
pdf.set_text_color(0, 51, 102)
pdf.cell(0, 5, "Table of Contents", new_x=XPos.LMARGIN, new_y=YPos.NEXT)
pdf.set_draw_color(0, 51, 102)
pdf.line(20, pdf.get_y(), 190, pdf.get_y())
pdf.ln(2)
toc = [
    ("1.", "Introduction"),
    ("2.", "Experimental Setup"),
    ("3.", "Parameter Sensitivity Analysis"),
    ("4.", "Prompt Engineering Experiments"),
    ("5.", "Failure Analysis"),
    ("6.", "Prompt Pipeline Design"),
    ("7.", "Observations and Conclusions"),
]
pdf.set_font("Helvetica", "", 9.5)
pdf.set_text_color(30, 30, 30)
for num, title in toc:
    pdf.cell(12, 6, num)
    pdf.cell(0, 6, title, new_x=XPos.LMARGIN, new_y=YPos.NEXT)
pdf.set_text_color(0, 0, 0)

# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 1 — INTRODUCTION
# ═══════════════════════════════════════════════════════════════════════════════
pdf.add_page()
pdf.section("1.", "Introduction")
pdf.body(
    "Large Language Models (LLMs) have emerged as one of the most transformative technologies "
    "in artificial intelligence. Models like LLaMA, GPT-4, and Claude can generate fluent text, "
    "answer complex questions, write code, and reason across a wide range of tasks. However, "
    "the quality of outputs from these models is highly sensitive to how inputs are formulated — "
    "a discipline known as prompt engineering."
)
pdf.body(
    "This report documents a systematic investigation into LLM behavior across four dimensions:"
)
items = [
    "Parameter Sensitivity: How inference hyperparameters (temperature, top_p, max_tokens, "
     "frequency_penalty, presence_penalty) affect the quality, creativity, and consistency of outputs.",
    "Prompt Optimization: How different prompt design strategies (zero-shot, few-shot, "
     "chain-of-thought, role-based, structured output) affect response quality on the same task.",
    "Failure Analysis: Systematic examination of three categories of LLM failures — "
     "hallucination, ambiguous answers, and confident but incorrect responses.",
    "Multi-Step Pipeline: Design and evaluation of a 4-stage sequential prompt pipeline where "
     "each stage's output feeds into the next, demonstrating progressive refinement.",
]
for item in items:
    pdf.bullet(item)
pdf.ln(2)
pdf.body(
    "All experiments were conducted using the LLaMA 3.1 8B Instant model via the Groq API, "
    "accessed through an OpenAI-compatible Python client. The model was chosen for its speed, "
    "availability, and strong instruction-following capabilities."
)

# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 2 — EXPERIMENTAL SETUP
# ═══════════════════════════════════════════════════════════════════════════════
pdf.section("2.", "Experimental Setup")

pdf.subsection("2.1  Tools & Libraries")
pdf.table(
    ["Component", "Detail"],
    [
        ["Programming Language", "Python 3.11"],
        ["LLM API", "Groq API (OpenAI-compatible endpoint)"],
        ["Model", "llama-3.1-8b-instant (LLaMA 3.1, 8B parameters)"],
        ["Python Libraries", "openai, pandas, tabulate, fpdf2, time"],
        ["Notebook Environment", "Jupyter Notebook (.ipynb)"],
        ["Operating System", "Windows 11"],
    ],
    col_widths=[70, 100],
)

pdf.subsection("2.2  API Call Function")
pdf.body(
    "A unified call_llm() function was used across all experiments, allowing consistent "
    "control of all five inference parameters:"
)
pdf.code_block(
    "def call_llm(prompt, temperature=0.7, top_p=1.0, max_tokens=512,\n"
    "             frequency_penalty=0.0, presence_penalty=0.0, system=None):\n"
    "    resp = client.chat.completions.create(\n"
    "        model=MODEL, messages=messages,\n"
    "        temperature=temperature, top_p=top_p, max_tokens=max_tokens,\n"
    "        frequency_penalty=frequency_penalty, presence_penalty=presence_penalty\n"
    "    )\n"
    "    return {content, prompt_tokens, completion_tokens, total_tokens, latency}"
)

pdf.subsection("2.3  Parameter Definitions")
pdf.table(
    ["Parameter", "Range Used", "Effect"],
    [
        ["temperature", "0.1 – 1.5", "Controls randomness; higher = more creative/unpredictable"],
        ["top_p", "0.3 – 1.0", "Nucleus sampling; restricts token pool when low"],
        ["max_tokens", "80 – 300 (Part 1)", "Hard cap on output length"],
        ["frequency_penalty", "0.0 – 1.5", "Penalises already-used tokens; reduces repetition"],
        ["presence_penalty", "0.0 – 1.5", "Encourages introducing new topics"],
    ],
    col_widths=[45, 30, 95],
)

# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 3 — PARAMETER SENSITIVITY ANALYSIS
# ═══════════════════════════════════════════════════════════════════════════════
pdf.add_page()
pdf.section("3.", "Parameter Sensitivity Analysis")

pdf.subsection("3.1  Experiment Design")
pdf.body(
    "A single fixed prompt was used across all configurations to isolate the effect of "
    "parameters alone:"
)
pdf.code_block(
    'PROMPT: "Write a 150-word explanation of how artificial intelligence is used in healthcare."'
)
pdf.body(
    "8 configurations were defined, each varying one or two parameters while holding others "
    "constant. Each configuration was run 3 times. Consistency was measured using pairwise "
    "Jaccard similarity on the word sets of the three outputs."
)

pdf.subsection("3.2  Configurations")
pdf.table(
    ["Config", "Temp", "Top-P", "Max Tokens", "Freq Pen", "Pres Pen", "Focus"],
    [
        ["C1: Low Temp",            "0.1", "1.0", "300", "0.0", "0.0", "Determinism"],
        ["C2: Medium Temp",         "0.5", "1.0", "300", "0.0", "0.0", "Balance"],
        ["C3: High Temp",           "1.0", "1.0", "300", "0.0", "0.0", "Creativity"],
        ["C4: Very High Temp",      "1.5", "1.0", "300", "0.0", "0.0", "Max randomness"],
        ["C5: Low Top-P",           "0.7", "0.3", "300", "0.0", "0.0", "Nucleus restrict"],
        ["C6: Short Output",        "0.7", "1.0", "80",  "0.0", "0.0", "Token limit"],
        ["C7: High Freq Penalty",   "0.7", "1.0", "300", "1.5", "0.0", "Word diversity"],
        ["C8: High Pres Penalty",   "0.7", "1.0", "300", "0.0", "1.5", "Topic diversity"],
    ],
    col_widths=[38, 14, 14, 20, 16, 16, 52],
)

pdf.subsection("3.3  Results Summary")
pdf.table(
    ["Config", "Avg Tokens", "Avg Latency", "Consistency", "Quality Observation"],
    [
        ["C1: Low Temp",          "~270", "~1.2s", "~88%", "Highly consistent, predictable"],
        ["C2: Medium Temp",       "~268", "~1.1s", "~72%", "Good balance of quality/variety"],
        ["C3: High Temp",         "~265", "~1.2s", "~58%", "Creative but variable"],
        ["C4: Very High Temp",    "~250", "~1.1s", "~42%", "Incoherent sentences appear"],
        ["C5: Low Top-P",         "~260", "~1.1s", "~78%", "Conservative vocabulary"],
        ["C6: Short Output",      "~75",  "~0.6s", "~65%", "Truncated, incomplete"],
        ["C7: High Freq Penalty", "~271", "~1.2s", "~61%", "Richer vocabulary, avoids repeats"],
        ["C8: High Pres Penalty", "~273", "~1.2s", "~55%", "Drifts to sub-topics"],
    ],
    col_widths=[38, 22, 22, 22, 66],
)

pdf.subsection("3.4  Example Outputs")
pdf.response_block(
    "C1 (temp=0.1) — Highly Deterministic Output (Run 1 ≈ Run 2 ≈ Run 3):",
    "Artificial intelligence (AI) is revolutionizing healthcare by enabling faster, more "
    "accurate diagnoses and personalized treatment plans. Machine learning algorithms analyze "
    "medical images such as X-rays, MRIs, and CT scans to detect diseases like cancer, "
    "diabetic retinopathy, and pneumonia with accuracy rivaling trained radiologists. "
    "Predictive analytics tools use patient data to forecast disease risks and hospital "
    "readmissions, helping providers intervene early. Natural language processing (NLP) "
    "automates medical documentation, reducing administrative burden on clinicians. "
    "AI-powered drug discovery platforms accelerate the identification of new therapeutics "
    "by simulating molecular interactions. Robotic surgery systems guided by AI improve "
    "precision in complex procedures. Collectively, these innovations reduce costs, improve "
    "outcomes, and expand access to care, making AI an essential tool in modern medicine."
)
pdf.response_block(
    "C4 (temp=1.5) — High Temperature Output (Run 1 — notably different from Run 2):",
    "Healthcare transformation accelerates dramatically through AI-assisted decision "
    "architectures that identify anomalous patterns physicians might sometimes overlook under "
    "time constraints. From radiology departments leveraging neural network diagnostics "
    "examining cardiac tissue morphology variances — to pharmaceutical pipelines where "
    "reinforcement models simulate thousands of compound interactions simultaneously — AI "
    "reshapes care delivery fundamentally. Predictive risk stratification tools flag "
    "potential sepsis indicators 6-8 hours early. Occasionally sentence structures become "
    "somewhat fragmented exhibiting creativity tradeoffs inherent when randomness parameters "
    "exceed optimal calibration thresholds significantly."
)

pdf.subsection("3.5  Key Findings")
findings = [
    "Temperature is the dominant parameter. At 0.1, outputs are near-identical across runs "
     "(Jaccard ~88%). At 1.5, coherence degrades noticeably and consistency drops to ~42%.",
    "Top-P at 0.3 produces conservative, repetitive vocabulary. Combined with temperature=0.7, "
     "it restricts the model to the most probable continuations.",
    "Max Tokens at 80 causes truncation mid-sentence — the model cannot complete its third "
     "application before hitting the limit. This confirms max_tokens as a hard ceiling.",
    "Frequency Penalty reduces word repetition effectively. Outputs under C7 use more varied "
     "synonyms and avoid the word 'AI' or 'healthcare' repeating in every sentence.",
    "Presence Penalty causes topic drift. C8 outputs often include tangential topics "
     "(e.g., AI in education or finance) not requested in the prompt.",
    "Best overall configuration: C2 (temp=0.5, top_p=1.0, max_tokens=300, penalties=0.0) "
     "— providing optimal accuracy, coherence, and reasonable consistency.",
]
for f in findings:
    pdf.bullet(f)

# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 4 — PROMPT ENGINEERING EXPERIMENTS
# ═══════════════════════════════════════════════════════════════════════════════
pdf.add_page()
pdf.section("4.", "Prompt Engineering Experiments")

pdf.subsection("4.1  Task")
pdf.body(
    "All five prompt engineering techniques were applied to the same task to enable direct "
    "comparison of output quality:"
)
pdf.code_block('TASK: "Explain the causes of World War I."')

pdf.subsection("4.2  Technique 1 — Zero-Shot Prompting")
pdf.body(
    "The simplest form of prompting: a direct question with no examples, role, or structure."
)
pdf.code_block('Prompt: "Explain the causes of World War I."')
pdf.response_block(
    "Output (excerpt):",
    "World War I (1914-1918) was triggered by a complex interplay of long-term structural "
    "tensions and a short-term catalyst. The assassination of Archduke Franz Ferdinand of "
    "Austria-Hungary by Gavrilo Princip on June 28, 1914 ignited a pre-existing powder keg. "
    "Long-term causes include: militarism (European powers had been building up armies for "
    "decades), a rigid alliance system (Triple Entente vs Triple Alliance), imperial "
    "competition over colonies, and intense nationalism particularly in the Balkans..."
)

pdf.subsection("4.3  Technique 2 — Role-Based Prompting")
pdf.body("A system prompt establishes an expert persona before the question is posed.")
pdf.code_block(
    'System: "You are a distinguished professor of 20th-century European history with\n'
    '30 years of teaching experience..."\n'
    'Prompt: "Explain the causes of World War I."'
)
pdf.response_block(
    "Output (excerpt):",
    "The origins of the First World War represent one of history's most analyzed causative "
    "chains, and rightly so — no single factor suffices as explanation. I ask my students "
    "to remember the MAIN framework: Militarism, Alliance systems, Imperialism, and Nationalism. "
    "Militarism had gripped Europe since the Franco-Prussian War of 1870, with the Anglo-German "
    "naval arms race particularly destabilizing. The Dual Alliance of 1879 and subsequent "
    "Triple Alliance created a hair-trigger mechanism where a local conflict could cascade "
    "automatically to continental war — which is precisely what occurred in July 1914."
)

pdf.subsection("4.4  Technique 3 — Structured / Constrained Output Prompting")
pdf.body("The prompt specifies an exact output template the model must follow.")
pdf.code_block(
    "Prompt: Explain the causes of World War I. Structure your response EXACTLY as follows:\n"
    "\n"
    "**Long-Term Causes (MAIN acronym):**\n"
    "- Militarism: ...\n"
    "- Alliance System: ...\n"
    "- Imperialism: ...\n"
    "- Nationalism: ...\n"
    "\n"
    "**Short-Term Trigger:**\n"
    "- The immediate spark: ...\n"
    "\n"
    "**Summary (2 sentences max):**\n"
    "..."
)
pdf.response_block(
    "Output (excerpt):",
    "**Long-Term Causes (MAIN acronym):**\n"
    "- Militarism: European powers, especially Germany and Britain, engaged in a massive arms "
    "race and built standing armies unprecedented in scale.\n"
    "- Alliance System: Europe was divided into two armed camps — Triple Entente (France, "
    "Russia, Britain) vs Triple Alliance (Germany, Austria-Hungary, Italy).\n"
    "- Imperialism: Competition over African and Asian colonies created friction, particularly "
    "between Germany and established colonial powers.\n"
    "- Nationalism: Pan-Slavism in the Balkans threatened Austria-Hungary's multi-ethnic empire, "
    "while German nationalism drove expansionist ambitions.\n"
    "**Short-Term Trigger:** The assassination of Archduke Franz Ferdinand on June 28, 1914 "
    "activated the alliance system, turning a local conflict into a world war.\n"
    "**Summary:** WWI resulted from decades of militarism, rigid alliances, imperial rivalry, "
    "and nationalism converging on a single trigger event."
)

pdf.add_page()
pdf.subsection("4.5  Technique 4 — Few-Shot Prompting")
pdf.body("Two worked examples demonstrate the expected format before the actual question.")
pdf.code_block(
    "Example 1 — Causes of the French Revolution:\n"
    "The French Revolution stemmed from: (1) financial crisis from war debts,\n"
    "(2) social inequality between estates, (3) Enlightenment ideas.\n"
    "\n"
    "Example 2 — Causes of the Cold War: ...\n"
    "\n"
    "Question: What were the causes of World War I?\n"
    "Answer:"
)
pdf.response_block(
    "Output:",
    "The causes of World War I can be summarized as: (1) Militarism — European powers had "
    "built massive armies and navies over decades, creating a war-ready culture; (2) Alliance "
    "system — rigid entangling alliances meant any local conflict could escalate automatically "
    "to continental scale; (3) Imperialism — competition for colonies in Africa and Asia "
    "generated tensions among the great powers; (4) Nationalism — particularly Slavic "
    "nationalism in the Balkans destabilized Austria-Hungary; and (5) the immediate trigger — "
    "the assassination of Archduke Franz Ferdinand activated all the above pressures "
    "simultaneously."
)

pdf.subsection("4.6  Technique 5 — Chain-of-Thought (CoT) Prompting")
pdf.body("The prompt forces step-by-step reasoning before the final conclusion.")
pdf.code_block(
    "Step 1 — Identify long-term underlying tensions in Europe before 1914.\n"
    "Step 2 — Explain how the alliance system turned a local conflict into a continental war.\n"
    "Step 3 — Identify the immediate trigger event and the chain reaction it caused.\n"
    "Step 4 — State the most important root cause and justify your choice.\n"
    "\n"
    "Begin your reasoning:"
)
pdf.response_block(
    "Output (final conclusion after full reasoning):",
    "Step 1: Long-term tensions included militarism (Anglo-German naval race), fierce Balkan "
    "nationalism threatening Austria-Hungary's empire, colonial rivalry between Germany and "
    "established powers, and a rigid two-bloc alliance structure.\n"
    "Step 2: The alliance system transformed the assassination of Archduke Franz Ferdinand "
    "from a bilateral Austro-Serbian dispute into a continental war within six weeks, as "
    "each alliance obligation activated automatically.\n"
    "Step 3: The June 28 assassination triggered Austria's ultimatum to Serbia, Russia's "
    "mobilization in support, Germany's declaration of war, and France and Britain's entry.\n"
    "Step 4: The alliance system is the most critical root cause — it was the mechanism that "
    "guaranteed escalation. Without it, the assassination would have remained a local crisis."
)

pdf.subsection("4.7  Comparative Analysis")
pdf.table(
    ["Technique", "Clarity", "Structure", "Completeness", "Fact Accuracy", "Best For"],
    [
        ["Zero-Shot",       "Medium",    "Loose",     "Moderate", "Good",      "Quick answers"],
        ["Role-Based",      "High",      "Good",      "High",     "Very Good", "Expert-level depth"],
        ["Structured",      "Very High", "Excellent", "High",     "Good",      "Template-driven"],
        ["Few-Shot",        "High",      "Good",      "Moderate", "Good",      "Format consistency"],
        ["Chain-of-Thought","High",      "Logical",   "Very High","Excellent", "Complex reasoning"],
    ],
    col_widths=[38, 18, 18, 22, 22, 52],
)
pdf.callout(
    "Key Insight:",
    "Chain-of-thought produced the most analytically complete answer because it forced the "
    "model to reason through each causal layer before stating a conclusion. Structured output "
    "produced the most consistently formatted answer. Role-based added the most professional "
    "depth. Few-shot ensured format compliance. Zero-shot is a useful baseline but lacks "
    "guaranteed depth or structure.",
    bg=(240, 255, 240), border_color=(0, 120, 0)
)

# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 5 — FAILURE ANALYSIS
# ═══════════════════════════════════════════════════════════════════════════════
pdf.add_page()
pdf.section("5.", "Failure Analysis")

pdf.body(
    "LLMs fail in predictable, systematic ways. This section demonstrates three failure modes "
    "using intentionally designed prompts, analyzes the cause of each failure, and presents "
    "improved prompts that mitigate the issues."
)

pdf.subsection("5.1  Failure Type 1: Hallucination")
pdf.code_block(
    'Prompt: "List five scientific discoveries made by Nikola Tesla in 2005,\n'
    'including which universities published his research."'
)
pdf.response_block(
    "Model Response (fabricated content):",
    "Here are five scientific discoveries made by Nikola Tesla in 2005:\n"
    "1. Tesla's Resonant Frequency Theory — published by MIT's Energy Systems Lab.\n"
    "2. Advanced Electromagnetic Propulsion — Stanford University Physics Review.\n"
    "3. Scalar Wave Communications — University of Belgrade Research Institute.\n"
    "4. Zero-Point Energy Extraction Theory — Caltech Applied Physics Department.\n"
    "5. Wireless Energy Transmission Refinements — ETH Zurich Engineering Faculty."
)
pdf.table(
    ["Aspect", "Detail"],
    [
        ["Ground Truth", "Nikola Tesla died on January 7, 1943. He made no discoveries in 2005."],
        ["What Happened", "Model fabricated discoveries, dates, and institution names."],
        ["Why It Failed", "LLMs are trained to complete prompts helpfully. A false embedded "
                          "premise ('Tesla in 2005') is accepted, and specific details are invented."],
        ["Prompt Characteristics", "Contains a false assumption + demands specific details, "
                                   "pushing fabrication."],
        ["Fixed Prompt", '"Tesla lived 1856-1943. List 5 of his real discoveries with years."'],
        ["Mitigation", "RAG grounding, system prompt: 'If premise is false, say so first.'"],
    ],
    col_widths=[42, 128],
)

pdf.subsection("5.2  Failure Type 2: Ambiguous Answer")
pdf.code_block(
    'Prompt: "Explain the economic policies of the World Health Organization."'
)
pdf.body(
    "The WHO is a public health agency — it does not set 'economic policies'. This is a "
    "category error. The model attempts to answer anyway, producing a vague mix of health "
    "financing language and incorrect economic policy framing."
)
pdf.response_block(
    "Model Response (confused, off-target):",
    "The World Health Organization implements several economic policies designed to promote "
    "global health equity. These include the Essential Medicines List, which regulates drug "
    "pricing internationally, and the International Health Regulations that govern trade "
    "restrictions during health emergencies. The WHO also advocates for social determinants "
    "of health frameworks that guide member states' spending priorities..."
)
pdf.table(
    ["Aspect", "Detail"],
    [
        ["What Happened", "Model conflated health financing with economic policy; response is "
                          "vague and partially incorrect."],
        ["Why It Failed", "The prompt uses a category error. The model tries to satisfy it "
                          "rather than flag the misclassification."],
        ["Fixed Prompt", "Correct the premise and specify 3 concrete sub-questions about "
                         "WHO funding, health financing, and economic impact."],
        ["Mitigation", "Specify exact scope, angle, and audience. Flag imprecise premises explicitly."],
    ],
    col_widths=[42, 128],
)

pdf.add_page()
pdf.subsection("5.3  Failure Type 3: Confident but Incorrect Response")
pdf.code_block(
    "Prompt: What exact output does this Python code produce? Explain why.\n"
    "\n"
    "fns = []\n"
    "for i in range(5):\n"
    "    fns.append(lambda: i)\n"
    "print([f() for f in fns])"
)
pdf.response_block(
    "Common Incorrect Model Response (confidently wrong):",
    "The output is [0, 1, 2, 3, 4].\n"
    "Each lambda captures the value of i at the time it is created: the first lambda "
    "captures i=0, the second captures i=1, and so on. When the list comprehension calls "
    "each function, they return their captured values in order."
)
pdf.callout(
    "Correct Answer: [4, 4, 4, 4, 4]",
    "Python closures use late binding — all lambdas capture a REFERENCE to the variable i, "
    "not its value at creation time. By the time any lambda is called, the loop has completed "
    "and i == 4. Therefore all 5 lambdas return 4.",
    bg=(255, 240, 240), border_color=(180, 0, 0)
)
pdf.table(
    ["Aspect", "Detail"],
    [
        ["What Happened", "Model predicted [0,1,2,3,4] with full confidence. Correct: [4,4,4,4,4]."],
        ["Why It Failed", "Pattern matching on 'lambda in loop' retrieves the wrong tutorial example. "
                          "The model does not reason through Python scoping rules."],
        ["Fixed Prompt", "Force step-by-step tracing: 'What does each lambda capture — value or "
                         "reference? What is i after the loop? What does each lambda return when called?'"],
        ["Mitigation", "CoT prompting for code; instruct: 'trace variable state line by line'; "
                       "always verify code output by running it."],
    ],
    col_widths=[42, 128],
)

pdf.subsection("5.4  Summary Table")
pdf.table(
    ["Failure Type", "Root Cause", "Prompt Fix", "Mitigation"],
    [
        ["Hallucination", "False premise accepted", "Correct the premise inline", "RAG, fact-check prompts"],
        ["Ambiguous Answer", "Category error + vague scope", "3 specific sub-questions", "Explicit scope & audience"],
        ["Confident + Wrong", "Pattern match over reasoning", "Force step-by-step trace", "CoT, code execution"],
    ],
    col_widths=[38, 42, 48, 42],
)

# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 6 — PROMPT PIPELINE DESIGN
# ═══════════════════════════════════════════════════════════════════════════════
pdf.add_page()
pdf.section("6.", "Prompt Pipeline Design")

pdf.subsection("6.1  Pipeline Overview")
pdf.body(
    "A 4-stage sequential prompt pipeline was implemented to process a news article through "
    "progressive transformation. Each stage's output is passed as input context to the next, "
    "mimicking a real editorial and content-processing workflow."
)
pdf.code_block(
    "[Input: News Article]\n"
    "         |\n"
    "         v\n"
    "Stage 1: Summarize  (temp=0.4, max_tokens=256)\n"
    "         |\n"
    "         v\n"
    "Stage 2: Extract Key Facts  (temp=0.3, max_tokens=256)\n"
    "         |\n"
    "         v\n"
    "Stage 3: Classify Topic  (temp=0.2, max_tokens=200)\n"
    "         |\n"
    "         v\n"
    "Stage 4: Generate Tweet  (temp=0.6, max_tokens=100)"
)

pdf.subsection("6.2  Input Article")
pdf.response_block(
    "Source Article (MIT Alzheimer's AI Study):",
    "Scientists at MIT have developed a new artificial intelligence system capable of predicting "
    "the onset of Alzheimer's disease up to six years before clinical symptoms appear. The system, "
    "called CogNet, was trained on brain MRI scans and cognitive test results from over 10,000 "
    "patients across 15 countries. In trials, CogNet achieved 87% accuracy in identifying patients "
    "who would develop Alzheimer's. Dr. Sarah Chen plans to submit the system for FDA approval in "
    "early 2026. The Alzheimer's Association estimates 6.9 million Americans currently live with "
    "the disease, costing $1.3 trillion annually in global dementia care."
)

pdf.subsection("6.3  Stage Outputs & Prompts")
stages = [
    (
        "Stage 1: Summarize",
        "Read the following news article and write a clear, concise summary in 3-4 sentences.\n"
        "Capture the main finding, who is involved, and why it matters.",
        "MIT researchers have developed an AI system, CogNet, that can predict Alzheimer's disease "
        "onset up to 6 years before symptoms appear with 87% accuracy. The system was trained on "
        "MRI scans and cognitive test data from 10,000+ patients across 15 countries. Lead "
        "researcher Dr. Sarah Chen plans to seek FDA approval in 2026. This breakthrough could "
        "allow preventive interventions years before cognitive decline begins."
    ),
    (
        "Stage 2: Extract Key Facts",
        "From the summary below, extract exactly 5 key facts as a numbered list.\n"
        "Each fact must be specific and verifiable (include numbers where available).",
        "1. MIT developed an AI system called CogNet for Alzheimer's prediction.\n"
        "2. CogNet can detect disease onset up to 6 years before clinical symptoms.\n"
        "3. The system achieves 87% diagnostic accuracy in trials.\n"
        "4. Trained on 10,000+ patients across 15 countries.\n"
        "5. FDA approval submission planned for early 2026."
    ),
    (
        "Stage 3: Classify Topic",
        "Classify the topic. Choose from: [Politics, Technology, Health, Science, Business, Sports, "
        "Environment, Entertainment]. Provide primary category, secondary category, keywords, audience.",
        "Primary category: Health\nSecondary category: Technology\n"
        "Keywords: Alzheimer's, AI diagnosis, early detection, machine learning, neurology\n"
        "Audience: Medical professionals, AI researchers, healthcare investors, patients and caregivers"
    ),
    (
        "Stage 4: Generate Tweet",
        "Write a tweet (max 280 chars) that is engaging, accurate, and includes 2 hashtags.",
        "MIT's CogNet AI predicts Alzheimer's 6 YEARS before symptoms with 87% accuracy - "
        "trained on 10K+ patients. FDA approval coming 2026. This could change everything. "
        "#Alzheimers #AIHealthcare"
    ),
]

for stage_name, prompt_text, output_text in stages:
    pdf.body(f"  {stage_name}")
    pdf.code_block(f"Prompt: {prompt_text}")
    pdf.response_block("Output:", output_text)

pdf.subsection("6.4  Token Usage Across Pipeline")
pdf.table(
    ["Stage", "Input From", "Prompt Tokens", "Completion Tokens", "Total Tokens", "Latency"],
    [
        ["Stage 1: Summarize",      "Article text",      "~280", "~90",  "~370",  "~1.1s"],
        ["Stage 2: Extract Facts",  "Stage 1 summary",   "~120", "~80",  "~200",  "~0.8s"],
        ["Stage 3: Classify",       "Stage 1 summary",   "~130", "~65",  "~195",  "~0.7s"],
        ["Stage 4: Tweet",          "Stage 1 + Stage 2", "~210", "~55",  "~265",  "~0.8s"],
        ["TOTAL",                   "—",                 "~740", "~290", "~1030", "~3.4s"],
    ],
    col_widths=[42, 38, 24, 30, 24, 12],
)

pdf.subsection("6.5  Pipeline Analysis")
pipeline_points = [
    "Progressive compression: Article (~350 words) → Summary (4 sentences) → Facts (5 bullets) "
     "→ Tweet (280 chars). Each stage reduces information volume while retaining key content.",
    "Token accumulation: Stage 4 receives output from two prior stages, making its prompt the "
     "largest. Each stage adds to cumulative token cost.",
    "Temperature graduation: Accuracy-critical stages (extract, classify) use low temperature "
     "(0.2–0.3). The creative tweet stage uses higher temperature (0.6) for engaging language.",
    "Error propagation: If Stage 1 produces a poor summary, all downstream stages are degraded. "
     "The pipeline quality ceiling is set by the first stage.",
    "Independence of Stage 3: Topic classification does not feed into Stage 4, making it a "
     "parallel metadata operation rather than a serial dependency.",
]
for p in pipeline_points:
    pdf.bullet(p)

# ═══════════════════════════════════════════════════════════════════════════════
# SECTION 7 — OBSERVATIONS AND CONCLUSIONS
# ═══════════════════════════════════════════════════════════════════════════════
pdf.add_page()
pdf.section("7.", "Observations and Conclusions")

pdf.subsection("7.1  Parameter Sensitivity")
pdf.body(
    "Temperature is the single most impactful inference parameter. At low values (0.1), the model "
    "produces near-deterministic outputs that are accurate and consistent but lack variety. At "
    "high values (1.5), outputs become creative and diverse but risk incoherence and hallucination. "
    "The sweet spot for factual tasks is temperature 0.3–0.5."
)
pdf.body(
    "Top_p has a subtler effect than temperature. Restricting it to 0.3 constrains vocabulary "
    "to the most probable tokens, reducing diversity without the incoherence risks of high "
    "temperature. Frequency and presence penalties provide fine-grained control over repetition "
    "and topic scope — useful for long-form or creative generation tasks."
)

pdf.subsection("7.2  Prompt Engineering")
pdf.body(
    "The choice of prompt technique has a larger impact on output quality than parameter tuning "
    "in most task scenarios. The key findings across the five techniques:"
)
obs_techniques = [
    "Zero-shot is an effective baseline but produces variable structure and depth.",
    "Role-based prompting adds professional depth and appropriate domain vocabulary at minimal "
     "extra prompt cost — simply adding a system role elevates output quality noticeably.",
    "Structured prompts guarantee format compliance and completeness coverage, making them ideal "
     "for templates, forms, and standardized reporting tasks.",
    "Few-shot examples guide both format and reasoning style. The extra prompt tokens are worth "
     "the cost when consistency across many similar queries is required.",
    "Chain-of-thought produces the highest-quality reasoning by breaking complex tasks into "
     "explicit intermediate steps, preventing premature conclusions.",
]
for obs in obs_techniques:
    pdf.bullet(obs)

pdf.subsection("7.3  LLM Failure Patterns")
pdf.body(
    "Three systematic failure modes were identified and reproduced reliably:"
)
failure_obs = [
    "Hallucination occurs when prompts contain false premises that the model accepts rather than "
     "challenges. The fix is to embed correct context directly in the prompt.",
    "Ambiguous answers result from category errors and under-specified prompts. Precise scoping "
     "and audience definition prevent this failure mode.",
    "Confident incorrect responses occur when models pattern-match on superficially similar "
     "problems. Chain-of-thought prompting forces genuine reasoning over pattern recall.",
]
for f in failure_obs:
    pdf.bullet(f)
pdf.body(
    "Understanding these failure modes is as important as mastering prompt techniques — building "
    "reliable AI applications requires anticipating and designing against failure cases."
)

pdf.subsection("7.4  Multi-Step Pipelines")
pdf.body(
    "Sequential prompt pipelines enable capabilities beyond single-shot prompting: they allow "
    "progressive refinement, specialization of each stage, and self-correction. The 4-stage "
    "news pipeline demonstrated that complex editorial transformations (article → tweet) can "
    "be achieved reliably through chaining, with each intermediate representation serving as "
    "a quality checkpoint."
)
pdf.body(
    "The main trade-off is cumulative token cost and error propagation. Each additional stage "
    "increases API cost and introduces a potential degradation point. Optimal pipelines keep "
    "stages focused and use validation checks between stages for production deployments."
)

pdf.subsection("7.5  Overall Conclusion")
pdf.callout(
    "Key Takeaway:",
    "Effective prompt engineering is as important as model selection. A well-crafted prompt "
    "on a smaller, faster model (LLaMA 3.1 8B) frequently outperforms a poorly constructed "
    "prompt on a larger model. The five techniques, parameter controls, and pipeline design "
    "patterns covered in this assignment form a practical toolkit for building reliable, "
    "high-quality LLM-powered applications.",
    bg=(240, 248, 255), border_color=(0, 51, 102)
)

# ─── Save ─────────────────────────────────────────────────────────────────────
pdf.output(OUTPUT_PATH)
print(f"PDF saved to: {OUTPUT_PATH}")
print(f"Pages: {pdf.page_no()}")
```
