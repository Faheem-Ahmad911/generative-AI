# Load Stanford's GloVe Embedding using API

https://radimrehurek.com/gensim/models/keyedvectors.html#gensim.models.keyedvectors.WordEmbeddingsKeyedVectors.most_similar

```python
import gensim.downloader as api
word_vectors = api.load("glove-wiki-gigaword-100")  # load pre-trained word-vectors from gensim-data
```

```python
# This method computes cosine similarity between a simple mean of the projection weight vectors of the
# word in the model. The method corresponds to the word-analogy and distance scripts in the original
# word2vec implementation. (positive=["woman", "king"], negative=["man"])
result = word_vectors.most_similar(positive=['woman', 'king'], negative=['man'])
print('{:.4f}'.format(result[0][1]))
```

```python
# Find the top-N most similar words, using the multiplicative combination objective, proposed by Omer
# Levy and Yoav Goldberg in "Linguistic Regularities in Sparse and Explicit Word Representations".
# Positive words still contribute positively towards the similarity, negative words negatively, but
# with less susceptibility to one large distance dominating the calculation.
result = word_vectors.most_similar_cosmul(positive=['woman', 'king'], negative=['man'])
print('{:.4f}'.format(result[0][1]))
```

```python
# Which document from the given list doesn't go with the others from the training set?
print(word_vectors.doesnt_match("breakfast cereal dinner lunch".split()))
```

```python
# Compute cosine similarity between two words
similarity = word_vectors.similarity('woman', 'man')
similarity > 0.8
```

```python
# Find the top-N most similar words
result = word_vectors.similar_by_word("cat")
print('{:.4f}'.format(result[0][1]))
```

```python
sentence_obama = 'Obama speaks to the media in Illinois'.lower().split()
sentence_president = 'The president greets the press in Chicago'.lower().split()
```

```python
# Compute the Word Mover's Distance between two documents
similarity = word_vectors.wmdistance(sentence_obama, sentence_president)
print('{:.4f}'.format(similarity))
```

```python
distance = word_vectors.distance("media", "media")
print('{:.1f}'.format(distance))
```

```python
# Compute cosine similarity between two sets of docvecs from the trained set.
sim = word_vectors.n_similarity(['sushi', 'shop'], ['japanese', 'restaurant'])
print('{:.4f}'.format(sim))
```
