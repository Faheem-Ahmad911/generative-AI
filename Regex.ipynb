```python
import re
import pandas as pd

# Define the 5 raw text samples for Question 1
raw_samples = [
    "<div>System_Log: User @Ali processed payment for 'Premium_Sub' on 2026-02-18. ID: TXN-5582. Status: Paid!</div>",
    "<div>System_Log: User @Zain attempted 'Basic_Plan' purchase. ID: TXN-9910. Error: Funds_Low. Status: Unpaid.</div>",
    "<div>System_Log: [IP: 192.168.1.1] User @Sana login successful. ID: TXN-2241. Status: Paid. Reference: http://logs.pucit.edu.pk/txn2241</div>",
    "<div>System_Log: User @Hamza requested refund for ID: TXN-4403. Verification: Pending. Status: Unpaid.</div>",
    "<div>System_Log: TXN-8827 confirmed for @User_99. Amount: 5000PKR. Status: Paid. (Verified)</div>"
]

def process_financial_logs(samples):
    results = []

    # Iterate through each raw text sample
    for i, text in enumerate(samples, 1):

        
        cleaning_pattern = r'</?div>|System_Log:\s*|\[IP:.*?\]\s*|Reference:\s*|http\S+|\s*\(Verified\)'

        # re.sub replaces the matched noise with an empty string
        cleaned_text = re.sub(cleaning_pattern, '', text).strip()


       
        tokens = re.findall(r'\b\w+\b', cleaned_text)


        
        extracted_txn = None
        txn_match = re.search(r'TXN-\d{4}', text) # Searching the original or cleaned text works
        if txn_match:
            extracted_txn = txn_match.group()


    
        status = "Unknown"
        if "Paid" in tokens:
            status = "Paid"
        elif "Unpaid" in tokens:
            status = "Unpaid"


        # Append the processed steps to our results list
        results.append({
            "Sample": f"Sample {i}",
            "Cleaning": cleaned_text,
            "Tokenization": str(tokens),
            "Extraction": extracted_txn,
            "Validation": status
        })

    return results

# Run the pipeline
pipeline_results = process_financial_logs(raw_samples)

# Display the results clearly
for res in pipeline_results:
    print(f"--- {res['Sample']} ---")
    print(f"Cleaning     : {res['Cleaning']}")
    print(f"Tokenization : {res['Tokenization']}")
    print(f"Extraction   : {res['Extraction']}")
    print(f"Validation   : {res['Validation']}")
    print("\n")

# Optional: If you want to visualize Sample 1 exactly like the requested table output
print("=== EXPECTED OUTPUT FORMAT FOR SAMPLE 1 ===")
df_sample1 = pd.DataFrame([
    {"Step": "Cleaning", "Expected Output": pipeline_results[0]["Cleaning"]},
    {"Step": "Tokenization", "Expected Output": pipeline_results[0]["Tokenization"]},
    {"Step": "Extraction", "Expected Output": pipeline_results[0]["Extraction"]},
    {"Step": "Validation", "Expected Output": pipeline_results[0]["Validation"]}
])
display(df_sample1) # 'display' works beautifully in Colab/Jupyter to render a table
```

```python
import re
import pandas as pd

# Define the 5 raw text samples for Question 2 [cite: 38-47]
raw_samples_q2 = [
    "<div>@Ali: Can't wait for the NLP_Summit on 2026-05-12! #Excited. Details at http://event.ai/join</div>",
    "<div>@Zain: Thinking about the workshop on 2026-06-15. #Interested. http://link.com</div>",
    "<div>@Sana: I will attend the 2026-07-20 meeting. #Neutral. No link.</div>",
    "<div>@Hamza: 2026-08-05 is the big day! #Excited to see you all. http://link.ai</div>",
    "<div>@Admin: Checking the 2
    026-09-10 calendar. #Interested. http://pucit.edu.pk</div>"
]

def process_social_media(samples):
    results = []

    # Iterate through each raw text sample [cite: 12]
    for i, text in enumerate(samples, 1):

        
        cleaning_pattern = r'</?div>|@\w+|http\S+'

        # We replace the noise with an empty string.
        # Note: The expected output for Sample 1 leaves the colon and spaces, so we just strip trailing whitespaces.
        cleaned_text = re.sub(cleaning_pattern, '', text).strip()

        
        tokens = re.findall(r'#\w+|\w+', cleaned_text)

        extracted_date = None
        date_match = re.search(r'\d{4}-\d{2}-\d{2}', text)
        if date_match:
            extracted_date = date_match.group()

        inference = "Unknown"
        for token in tokens:
            if token.startswith('#'):
                inference = token.replace('#', '')
                break

        # Append the processed steps to our results list
        results.append({
            "Sample": f"Sample {i}",
            "Cleaning": cleaned_text,
            "Tokenization": tokens, # Leaving as list to match expected output format
            "Extraction": extracted_date,
            "Inference": inference
        })

    return results

# Run the pipeline
pipeline_results_q2 = process_social_media(raw_samples_q2)

# Display the output for all 5 samples [cite: 12]
for res in pipeline_results_q2:
    print(f"--- {res['Sample']} ---")
    print(f"Cleaning     : {res['Cleaning']}")
    print(f"Tokenization : {res['Tokenization']}")
    print(f"Extraction   : {res['Extraction']}")
    print(f"Inference    : {res['Inference']}")
    print("\n")

# Replicate the specific table output for Sample 1 as requested in the assignment [cite: 48, 49]
print("=== EXPECTED OUTPUT FORMAT FOR SAMPLE 1 ===")
df_sample1_q2 = pd.DataFrame([
    {"Step": "Cleaning", "Expected Output": pipeline_results_q2[0]["Cleaning"]},
    {"Step": "Tokenization", "Expected Output": str(pipeline_results_q2[0]["Tokenization"])},
    {"Step": "Extraction", "Expected Output": pipeline_results_q2[0]["Extraction"]},
    {"Step": "Inference", "Expected Output": pipeline_results_q2[0]["Inference"]}
])

# Use display() to render it nicely if you are in Colab/Jupyter
display(df_sample1_q2)
```

```python
import re
import pandas as pd

# Define the 5 raw text samples for Question 3 [cite: 54-63]
raw_samples_q3 = [
    "<div>[08:30:15] LOG: User_A swiped at Entry_1. ID: BID-10293. Status: Authorized.</div>",
    "<div>[09:45:00] LOG: User_B attempted restricted area. ID: BID-44921. Status: Denied. Access Level: Low.</div>",
    "<div>[10:15:22] LOG: Guest_X check-in complete. ID: BID-00211. Status: Authorized. http://security.pucit.edu.pk/logs</div>",
    "<div>[12:00:10] LOG: User_C failed verification. ID: BID-99382. Status: Denied.</div>",
    "<div>[14:20:55] LOG: Maintenance_Z cleared for entry. ID: BID-33810. Status: Authorized. (Verified_Biometrically)</div>"
]

def process_security_logs(samples):
    results = []

    # Iterate through each raw text sample
    for i, text in enumerate(samples, 1):

        
        cleaning_pattern = r'</?div>|\[\d{2}:\d{2}:\d{2}\]\s*|http\S+|\s*\(Verified_Biometrically\)'

        cleaned_text = re.sub(cleaning_pattern, '', text).strip()

        tokens = re.findall(r'\w+', cleaned_text)


        extracted_bid = None
        bid_match = re.search(r'BID-\d{5}', text)
        if bid_match:
            extracted_bid = bid_match.group()

        validation = "Unknown"
        if "Authorized" in tokens:
            validation = "Authorized"
        elif "Denied" in tokens:
            validation = "Denied"

        # Append the processed steps to our results list
        results.append({
            "Sample": f"Sample {i}",
            "Cleaning": cleaned_text,
            "Tokenization": tokens,
            "Extraction": extracted_bid,
            "Validation": validation
        })

    return results

# Run the pipeline
pipeline_results_q3 = process_security_logs(raw_samples_q3)

# Display the output for all 5 samples
for res in pipeline_results_q3:
    print(f"--- {res['Sample']} ---")
    print(f"Cleaning     : {res['Cleaning']}")
    print(f"Tokenization : {res['Tokenization']}")
    print(f"Extraction   : {res['Extraction']}")
    print(f"Validation   : {res['Validation']}")
    print("\n")

# Replicate the specific table output for Sample 1 as requested in the assignment [cite: 64, 65]
print("=== EXPECTED OUTPUT FORMAT FOR SAMPLE 1 ===")
df_sample1_q3 = pd.DataFrame([
    {"Step": "Cleaning", "Expected Output": pipeline_results_q3[0]["Cleaning"]},
    {"Step": "Tokenization", "Expected Output": str(pipeline_results_q3[0]["Tokenization"])},
    {"Step": "Extraction", "Expected Output": pipeline_results_q3[0]["Extraction"]},
    {"Step": "Validation", "Expected Output": pipeline_results_q3[0]["Validation"]}
])

# Use display() to render the final table
display(df_sample1_q3)
```
