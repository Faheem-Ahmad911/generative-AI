```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, wordpunct_tokenize, TreebankWordTokenizer

# Download BOTH required NLTK data packages
nltk.download('punkt')
nltk.download('punkt_tab') # <-- This is the missing piece!

# Sample Corpus from the slides
corpus = """Hello! Welcome  to NLP Class.
Please do note the lectures, to become expert in NLP."""

print("--- Sentence Tokenization ---")
documents = sent_tokenize(corpus)
print(documents)

print("\n--- Word Tokenization ---")
words = word_tokenize(corpus)
print(words)
print("Punctuation")
punctuation=wordpunct_tokenize(corpus)

print(punctuation)
```

Stemmer and its Types

```python
import nltk
from nltk.stem import PorterStemmer, RegexpStemmer, SnowballStemmer

# 1. Initialize the three stemmers
porter_stemmer = PorterStemmer()
# RegexpStemmer takes a regular expression and removes matching prefixes/suffixes [cite: 271]
regexp_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
# Snowball Stemmer (Porter2) requires you to specify the language [cite: 287, 288]
snowball_stemmer = SnowballStemmer('english')

# 2. Words chosen specifically from your slides to highlight differences
test_words = [
    "eating",        # Standard verb ending [cite: 230]
    "goes",          # Irregular verb ending [cite: 314]
    "history",       # Ends with 'y' [cite: 231]
    "fairly",        # Adverb ending [cite: 308]
    "sportingly",    # Complex adverb ending [cite: 308]
    "ingeating"      # To test Regex prefix/suffix behavior [cite: 277]
]

# 3. Print the comparison
print(f"{'Original Word':<15} | {'Porter':<12} | {'Snowball':<12} | {'Regexp':<12}")
print("-" * 58)

for word in test_words:
    p_stem = porter_stemmer.stem(word)
    s_stem = snowball_stemmer.stem(word)
    r_stem = regexp_stemmer.stem(word)

    print(f"{word:<15} | {p_stem:<12} | {s_stem:<12} | {r_stem:<12}")
```

Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer

# 1. Download the WordNet dictionary (run once)
nltk.download('wordnet')

# 2. Initialize the Lemmatizer
lemmatizer = WordNetLemmatizer()

# ==========================================
# PART 1: Demonstrating all POS Types
# ==========================================
print("--- Lemmatization by POS Type ---")

# 1. Noun ('n') - This is the default if you don't specify a POS
print(f"Noun (programs)    -> {lemmatizer.lemmatize('programs', pos='n')}")

# 2. Verb ('v') - Converting an action to its base dictionary form
print(f"Verb (going)       -> {lemmatizer.lemmatize('going', pos='v')}")

# 3. Adjective ('a') - Finding the base degree of an adjective
print(f"Adjective (better) -> {lemmatizer.lemmatize('better', pos='a')}")

# 4. Adverb ('r') - Finding the base form of an adverb
print(f"Adverb (fairly)    -> {lemmatizer.lemmatize('fairly', pos='r')}")


# ==========================================
# PART 2: The Loop from your Lecture Slides
# ==========================================
# Your professor's specific word list
words = ["eating", "eats", "eaten", "writing", "writes",
         "programming", "programs", "history", "finally", "finalized"]

print("\n--- Lemmatizing a list of words (Foursome) ---")
```
